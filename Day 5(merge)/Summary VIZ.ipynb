## 1.Location Summary on MAP

```python
import folium
import pandas as pd
from IPython.display import display

# Load data from the CSV file
csv_file = r"C:\Users\Aditya\OneDrive\Desktop\WORK\Day 5(merge)\MAIN it 1.csv"  # Update with the path to your CSV file
df = pd.read_csv(csv_file)

# Define a list of unique areas
unique_areas = df['Area'].unique()

# Define a custom set of colors for each area
colors = ['red', 'green', 'blue', 'purple', 'orange', 'pink', 'brown', 'gray']

# Create a base map
m = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], zoom_start=5)

# Plot each point on the map with a color based on the area
for i, row in df.iterrows():
    area = row['Area']
    color = colors[list(unique_areas).index(area)]
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=6,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
    ).add_to(m)

# Display the map in the Jupyter Notebook
m.save("map.html")
display(m)
```

## 2.Location Summary PLOT

```python
import pandas as pd
import matplotlib.pyplot as plt



# Load data from the CSV file
csv_file =r"C:\Users\Aditya\OneDrive\Desktop\WORK\Day 5(merge)\MAIN it 1.csv"
df = pd.read_csv(csv_file)

# Define a list of unique areas
unique_areas = df['Area'].unique()

# Define a custom set of colors for each area
colors = ['red', 'green', 'blue', 'purple', 'orange', 'pink', 'brown', 'gray']

# Create a scatter plot with colored labels
plt.figure(figsize=(10, 8))
for i, area in enumerate(unique_areas):
    area_df = df[df['Area'] == area]
    plt.scatter(area_df['longitude'], area_df['latitude'], c=colors[i], label=area, s=20)

plt.title('Scatter Plot of Cities')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True)
plt.show()
```
