```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
expired_df = pd.read_csv('expired.csv')
```

```python
expired_df 
```

```python
# Count the number of renewed and not renewed packs
renewed_counts = expired_df['Renewed'].value_counts()
total_counts = renewed_counts.sum()

# Calculate percentages
percentages = (renewed_counts / total_counts) * 100

# Plotting the pie chart with custom colors
labels = ['Not Renewed', 'Renewed']
colors = ['#FF4136', '#FFFFFF', '#000000', '#2ECC40', '#FFDC00']  # Red, White, Black, Green, Yellow

plt.figure(figsize=(8, 6))
plt.pie(percentages, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)

# Add count labels with contrasting colors
for i, (label, percent) in enumerate(zip(labels, percentages)):
    count = renewed_counts[i]
    plt.text(0.5 * (1 - i), 0.5 * (1 - i), f'{label}\nCount: {count}', horizontalalignment='center', verticalalignment='center', fontsize=12, color='black' if i < 2 else 'white', fontweight='bold')

plt.title('Renewed vs Not Renewed Packs')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
